```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# data of images: iris flower, each list is a data
iris.data
```

```python
# output of data
iris.target
```

```python
x = iris.data
y = iris.target

feature_names = iris.feature_names
feature_names
# print(x.shape)

# type(x) # numPy Array
```

```python
target_names = iris.target_names
target_names
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.5)


print(X_train.shape)
print(X_test.shape)
```

```python
print("X_Train:", X_train)
print("X_Test:", X_test)
print("Y_Train:", Y_train)
print("Y_Test:", Y_test)
```

```python
# USING MODEL from sklearn

# from sklearn.neighbors import KNeighborsClassifier
# knn = KNeighborsClassifier(n_neighbors=3)

from sklearn.tree import DecisionTreeClassifier
knn = DecisionTreeClassifier()

knn.fit(X_train, Y_train)
# y_predict = knn.predict(X_test)
# y_predict
```

```python
Y_pred = knn.predict(X_test)
Y_pred
```

```python
Y_test
```

```python
len(Y_pred)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(Y_test,Y_pred))
```

```python
# give your own sample to model to predict
sample     = [[3,5,6,2], [4,4,4,2], [1,1,1,2]]
predictions = knn.predict(sample)
# predictions
pred_species = [iris.target_names[p] for p in predictions]
print("Predictions: ", pred_species)
```

```python
# Model Persistence
# from sklearn import joblib
import joblib
joblib.dump(knn, 'mlbrain.joblib')
# model
```

```python
# Using Model Persistance, we used saved model to predict sample data
# No training of model needed, .fit not needed
# stores train model in a binary file

model = joblib.load('mlbrain.joblib')
model.predict(X_test)

sample     = [[3,5,6,2], [4,4,4,2], [1,1,1,2]]
predictions = model.predict(sample)

pred_species = [iris.target_names[p] for p in predictions]
print("Predictions: ", pred_species)
```
